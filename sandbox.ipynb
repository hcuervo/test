```python
# Importar Librerias
import pandas as pd
import numpy as np

import datetime as datetime
```

```python
data = pd.read_csv('/Users/horaciocuervo/Downloads/Node-Red 1 - Sheet1.csv')
```

```python
data['date'] = pd.to_datetime(data['TimeStamp'],origin='unix',unit='ms')
data = data.drop(columns=['TimeStamp'])
```

```python
data.set_index('date',inplace=True)
data.columns
```

```python
p_10 = data['Internal'].quantile(0.01)
p_90 = data['Internal'].quantile(0.99)
data = data[data['Internal'].gt(p_10) & data['Internal'].lt(p_90)]
p_10 = data['External'].quantile(0.01)
p_90 = data['External'].quantile(0.99)
data = data[data['External'].gt(p_10) & data['External'].lt(p_90)]
```
